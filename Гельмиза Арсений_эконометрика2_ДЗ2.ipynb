#### Задание 1

##### Рассмотрим MA(2) процесс y_t = 4 + ε_t − 0.2ε_t − 1+ 0.8ε_t−2, где величины εt независимы и нормально распределены N(0;4).

##### (c) (4 балла) Сгенерируйте траекторию данного процесса длиной 50 наблюдений. Постройте график ряда, график первых десяти значений выборочной ACF и PACF. Прокомментируйте поведение выборочных ACF и PACF.

Построим график ряда для данного процесса длиной 50 наблюдений:

```python
np.random.seed(0)  
eps = np.random.normal(0, 2, 50)  
y=[]
yt = np.zeros(50)
for t in range(50):
    yt[t] = 4 + eps[t] - 0.2 * eps[t-1] + 0.8 * eps[t-2]
    y.append(yt)
```

```python
plt.figure(figsize=(10, 4))
plt.plot(range(50), yt, label="MA(2) процесс")
plt.xlabel("временной ряд")
plt.ylabel("значение")
plt.title('MA(2) процесс')
plt.legend()
plt.show()
```

Построим график первых десяти значений выборочной ACF:

```python
from statsmodels.tsa.stattools import acf
```

```python
y_array = np.array(y)
y_l = y_array.flatten()


plt.figure(figsize=(10, 4))
plt.plot(acf(y_l,nlags=10))
plt.xlabel("Lag")
plt.ylabel("ACF")
plt.title("ACF")

plt.show()
```

Построим график первых десяти значений выборочной PACF:

```python
from statsmodels.tsa.stattools import pacf
```

```python
plt.figure(figsize=(10, 4))
plt.plot(pacf(y_l,nlags=10))
plt.xlabel("lag")
plt.ylabel("PACF")
plt.title("PACF")

plt.show()
```

Вывод: Графики выборочных ACF и PACF показывают, что после первого лага значения автокорреляции снижаются и затем показывают рост при втором лаге. Это подверждает структуру ma(2) процесса.

#### Задание 2

##### Рассмотрим процесс случайного блуждания yt = 1 + yt−1 + εt, где величины εt независимы и нормально распределены N (0; 1), а y0 = 4.

##### (c) (4 балла) Сгенерируйте траекторию данного процесса длиной 50 наблюдений. Постройте график ряда, график первых десяти значений выборочной ACF и PACF. Прокомментируйте поведение выборочных ACF и PACF.

Построим график ряда для данного процесса длиной 50 наблюдений:

```python
np.random.seed(0)  
eps3 = np.random.normal(0, 2, 50)  
y3=[]
yt3 = np.zeros(50)
for t in range(50):
    yt3[t] = 1 + yt3[t-1] + eps3[t]
    y3.append(yt3)

plt.figure(figsize=(10, 4))
plt.plot(range(50), yt3, label="Процесс случайного блуждания")
plt.xlabel("временной ряд")
plt.ylabel("значение")
plt.title("Процесс случайного блуждания")
plt.legend()
plt.show()
```

Построим график первых десяти значений выборочной ACF:

```python
y_array3 = np.array(y3)
y_l3 = y_array3.flatten()


plt.figure(figsize=(10, 4))
plt.plot(acf(y_l3,nlags=10))
plt.xlabel("Lag")
plt.ylabel("ACF")
plt.title("ACF")

plt.show()
```

Построим график первых десяти значений выборочной PACF:

```python
y_array3 = np.array(y3)
y_l3 = y_array3.flatten()


plt.figure(figsize=(10, 4))
plt.plot(pacf(y_l3,nlags=10))
plt.xlabel("Lag")
plt.ylabel("PACF")
plt.title("PACF")

plt.show()
```

Вывод: Выборочная ACF не сходится к нулю, а равномерно убывает, что подтверждает структуру процесса случайного блуждания. Выборочная PACF при первом лаге имеет значение близкое к 1 и затем убывает с ростом лагов, что также свидетельствует, что процесс является процессом случайного блуждания.
